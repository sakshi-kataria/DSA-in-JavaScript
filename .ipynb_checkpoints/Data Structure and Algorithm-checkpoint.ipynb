## STACK Implementation

```javascript
function stack(){
    this._size=0;
    this._storage=[];
}

stack.prototype.peek=function(){
    return this._storage[this._size];
}

stack.prototype.isEmpty=function(){
    //console.log('this._size',this._size);
    if(this._size>0) 
        return true; 
    else 
        return false;
}


stack.prototype.push=function(data){
    this._size=this._size+1;
    this._storage[this._size]=data;
}

stack.prototype.pop=function(){
    var size=this._size,
        deleteData;
    deleteData=this._storage[this._size];
    //console.log('this.isEmpty()',this.isEmpty());
    if(this.isEmpty()){
        delete this._storage[size];
        this._size=this._size-1;
    }else{
        console.log('empty stack')
    }
    return deleteData;
}

var S=new stack();
S.push(1);
S.push(2);
S.push(3);
console.log(S.peek());
console.log(S._storage);
S.pop();
console.log(S._storage);
S.pop();
console.log(S._storage);
```

## Queue:

A Queue is different form stack in that its removal and insertion follow FIFO (fist in fist out) principle.

Element are inserted in rear and removed from front.

**Three Fundamental methods are: ***

   **New()**: create empty queue.
   
   **enqueue(S:ADT,o:Object)**: insert object o at the rear of Queue S.
   
   ** dequeue(S:ADT)**: Remove object from the front of Queue ; an error occured if the queue is empty.
   
   **front(S:ADT)** : Return , but does not remove, front element; An error occured if queue is empty.
   
** Other Methods **

   **SIZE(S:ADT)**
   
   **ISEmpty(S:ADT)**
   
## Queue Implemenation

```javascript
var queue=function(){
    this._newestIndex=0;
    this._oldestIndex=0;
    this.storage=[];
}
queue.prototype.enqueue=function(item){
    if(item){
            this.storage[this._newestIndex]=item;
            this._newestIndex=this._newestIndex+1;
        }
    else{
        console.log('data is not correct');
    }
}
queue.prototype.dequeue=function(){
    if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        var deletedItem;
        deletedItem =this.storage[0];
        delete this.storage[0];
        this._oldestIndex=this._oldestIndex+1;
        console.log('Item is popup from Queue is '+ deletedItem);
    }
}
queue.prototype.front=function(){
     if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        console.log('Item is on front Queue is '+ this.storage[this._oldestIndex]);
    }
}

queue.prototype.size=function(){
    return (this._newestIndex - this._oldestIndex)
}

queue.prototype.printData=function(){
    var firstIndex=this._oldestIndex;
    if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        while(firstIndex !== this._newestIndex){
            console.log('Item:',this.storage[firstIndex])
            firstIndex=firstIndex+1;
        }
    }
}

var queue=new queue();
queue.front();
queue.dequeue();
queue.enqueue(1);
queue.front();
queue.enqueue(2);
queue.enqueue(3);
queue.dequeue();
queue.front();
console.log('size',queue.size());
queue.printData();
```

## Dictionaries

It is Abstract Data type that store element so that they can be located quickly using keys.

example:-> Dictionary store bank account then key is Account Number.

    * Store key element pair
    * The main operation supported is search by key.

**Methods :**
        * Size()    : return size of Dictionary.
        * IsEmpty() : return Dictionary is empty or not
        * elements(): return elements of Dictionary.
        
** Query Methos **
        * findElement(k)   : return searched element in Dictionary
        * findAllElement(k): return all searched element in Dictionary

** Updated Method **
        * removeItem(k)    : remove element from Dictionary
        * removeAllItem(k) : remove all element from Dictionary
        * insertItem(k,v): Insert element in Dictionary.
        
 * NIL is returned by unSuccessful Search *
 

```javascript
function Dictionary(){
    this.insertItem=insertItem;
    this.dataStore=[];
    this.removeItem=removeItem;
    this.findElement=findElement;
    this.elements=elements;
    this.size=size;
    this.isEmpty=isEmpty;
    this.removeAllElements=removeAllElements;
}

function isEmpty(){
    if (Object.keys(this.dataStore).length>0)
        return false;
    else
        return true;
}

function insertItem(k,e){
    this.dataStore[k]=e;
}

function removeItem(k){
    if(this.dataStore.hasOwnProperty(k)){
        delete this.dataStore[k];
    }else{
        console.log('key is not available in dictionary')
    }
}

function findElement(k){
    if(this.dataStore.hasOwnProperty(k)){
        return dataStore[k];
    }else{
        console.log('key is not available in dictionary')
    }
}

function elements(){
    if(Object.keys(this.dataStore).length)
    {
         for (key in (this.dataStore)){
            console.log(key , '-->' , this.dataStore[key]);
        }
    }else{
        console.log('Empty Dictionary');
    }
}

function size(){
    var n=0;
    for (var key in this.dataStore){
        n=n+1;
    }
    return n;
}

function removeAllElements(){
    for(var key in this.dataStore){
        delete this.dataStore[key];
    }
}

var book=new Dictionary();
book.insertItem("Raymond","123"); 
book.insertItem("David", "345"); 
book.insertItem("Cynthia", "456");
book.elements();
book.removeItem("Raymond");
book.elements();
console.log('size:',book.size());
console.log('empty:',book.isEmpty());
book.removeAllElements();
book.elements();
console.log('size:',book.size());
console.log('empty:',book.isEmpty());
```

## Binary Search

**IDEA**: Divide and conqure key design technique.
Narrow down the search ranges in the stages.

<img src="BinarySearch.png",width=500,height=300>

#### Algorithm

Algorith BinarySearch(A,k,low,high)

    if(low>high) then return Nil;
    else 
        mid <- (low+high/2);
        if(k=A[mid]) return mid
        else if (k>A[mid])
            return BinarySearch(A,k,low,mid-1)
        else
            return BinarySearch(A,k,mid+1,high)
            
 #### Implementation in JavaScript
        

```javascript
function BinarySearch(A,data){
    high=A.length-1;
    low=0
    while(low<=high){
        var mid=Math.floor((low+high)/2);
//         console.log('A[mid]',A[mid]);
        if(data===A[mid]){
            return mid+1;
        }else if(data>=A[mid]){
           low=mid+1;
        }else{
            high=mid-1;
        }
    }
    return -1;
}

A=[1,2,3,4,55,66,456,999];
console.log('BinarySearch',BinarySearch(A,0))
```

Tree represent hierarchy for eg organization structure of corporation

```javascript
function Node(val){
    this.val=val;
    this.left=null;
    this.right=null;
    this.parent=null;
    }

    function bst(){
        this.root=null;
    }

    bst.prototype.push=function(val){
        // If tree is empty
        var n=new Node(val);
        if(this.root===null){
            this.root=n;
            return;
        }

        //If tree is not empty
        var current= this.root;
        while(current !=null){
            if(val<current.val){
                if(current.left!=null){
                    current=current.left;
                }else{
                    current.left=n;
                    n.parent=current;
                    return;
                }
            }else{
                 if(current.right!=null){
                    current=current.right;
                }else{
                    current.right=n;
                    n.parent=current;
                    return;
                }
            }
        }    
    }

    bst.prototype.inOrderTraversal=function(node){
        var current= node;
        if(current!=null){
//             console.log('current.left',current.left);
            if(current.left){
                this.inOrderTraversal(current.left);
            }
            console.log(current.val);
//            console.log('current.right',current.right);
            if(current.right){
                this.inOrderTraversal(current.right);
            }
        }
    }
    bst.prototype.preOrderTraversal=function(node){
        var current= node;
        if(current!=null){
            console.log(current.val);
            if(current.left){
                this.preOrderTraversal(current.left);
            }
            if(current.right){
                this.preOrderTraversal(current.right);
            }
        }
    }
    bst.prototype.postOrderTraversal=function(node){
        var current= node;
        if(current!=null){
            if(current.left){
                this.postOrderTraversal(current.left);
            }
            if(current.right){
                this.postOrderTraversal(current.right);
            }
            console.log(current.val);
        }
    }
    
    bst.prototype.searchNode=function(node){
        var current= this.root;
        while(current!=null){
            if(node==current.val){
                console.log('Node exists in tree');
                return;
            }
            if(node<current.val){
                if(current.left!=null){
                    current=current.left;
                }else{
                    console.log('Node does not exists');
                    return;
                }
            }else{
                 if(current.right!=null){
                    current=current.right;
                }else{
                    console.log('Node does not exists');
                    return;
                }
            }
            
        }
        console.log('Node does not exists');
    }
    
    bst.prototype.MaxNode=function(){
        var current= this.root;
        var maxVal=this.root.val;
        while(current!==null){
            if(current.right && maxVal<current.right.val){
                maxVal=current.right.val;
                current=current.right;
            }else{
                return maxVal
            }
        }
        return maxVal;
    }
    
     
    bst.prototype.MinNode=function(){
        var current= this.root;
        var minVal=this.root.val;
        while(current!=null){
            if(current.left && minVal > current.left.val){
                minVal=current.left.val;
                current=current.left;
            }else{
                return minVal
            }
        }
        return minVal;
    }
    
    bst.prototype.successorNode =function(node){
         var current= this.root;
         var SucNode=this.root.val;
         while(current!=null){
//              console.log('***********parent*********',current.parent)
            if(current.val===node){
                console.log('current.val',current.val)
                if(current.right===null){
                    while(current.parent!=null){
                          console.log('current.parent.val',current.parent.val)
                        if(current.parent.val > Node){
                             SucNode=current.parent.val;
                        }
                    }
                    return;
                }else{
                    while(current!=null){
                        console.log('current.left.val',current.left.val)
                        SucNode=current.left.val;
                        current=current.left;
                    }
                }
             }else {
                if(node<current.val){
                    current=current.left;
                }else{
                    current=current.right;
                }
             }
             
         }
        return SucNode;
    }
    
    
    bst.prototype.predecessorNode =function(node){
         var current= this.root.left;
         var preNode=this.root.left.val;
         while(current!=null){
            if(current.val===Node){
                if(current.right===null){
                    return;
                }else{
                    preNode=current.right.val;
                    current=current.right;
                }
             }else {
                 preNode=current
                 current=current.left
             }
         }
        return preNode;
    }
    
//     bst.prototype.delete=function(){
        
//     }
    
    var bst1=new bst();
    bst1.push(7);
    bst1.push(6);
    bst1.push(23);
    bst1.push(4);
    bst1.push(2);
    bst1.push(11);
    bst1.push(5);
    bst1.push(25);
    
     console.log('*****************In Order*****************')
    bst1.inOrderTraversal(bst1.root);
    console.log('*****************Pre Order*****************')
    bst1.preOrderTraversal(bst1.root);
    console.log('*****************Post Order*****************');
    bst1.postOrderTraversal(bst1.root);
    console.log('*****************Max Node*****************');
    console.log(bst1.MaxNode());
    console.log('*****************Min Node*****************');
    console.log(bst1.MinNode());
    console.log('***************** Search Node *****************');
    console.log(bst1.searchNode(5));
    console.log('*****************successor Node*****************');
    console.log(bst1.successorNode(5));
//
```
