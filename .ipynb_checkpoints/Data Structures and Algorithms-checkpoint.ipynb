#### Strategy:
* start empty handed
* Insert a card in the right position of already sorted hand
* Continue untillall card inserted/sorted

Output: `A` should be like `A[1]<=A[2]<=A[3]`

* 3 <- **1**
* 9 <- **i**
* 7 <- **j**
* 1 <- **n**

` for j <- 2 to n do
     key <- A[j]
     i <- j-1
     while i > 0 and A[i] > key
      do A[i+1] <- A[i]
         i--
     A[i+1] <- key`

````javascript
var arrForSorting=[3,4,6,8,9,7,2,5,1];
function InsertionSort(arr){
    for(var i=1;i<arr.length;i++){
        var key=arr[i];
        var j=i-1;
        for(;j>=0 && arr[j]>key;j--){
            arr[j+1]=arr[j];
        }
        arr[j+1]=key;
    }
}
console.log(arrForSorting);
InsertionSort(arrForSorting);
console.log(arrForSorting);
````

### Best/ Worst/ Average Case

**Best Case**: Elements already sorted; tj=1( I will check last element if it is less than it will reain same), running time=f(n), i.e. linear time.

**Worst Case**: Elements are sorted in inverse order, tj=j, running time=f(n^2); quadratic time.

**Average Case**: tj=j/2, running time=f(n^2); quadratic time.

# Asymptotatic Notation:


## The "Big-Oh" O-notation:

The notation Ο(n) is the formal way to express the upper bound of an algorithm's running time. It measures the worst case time complexity or the longest amount of time an algorithm can possibly take to complete.

Used for  worst case only.

f(n)=O(g(n)),

if there exists constant c and n0 (n not) then f(n)  <= cg(n) for n>n0.

**example**:

<img src="BigOhExample.png",width=400,height=200>

In this example 

f(n)=2n+6---> O(n)

    c1 g(n) < f(n)  < c2 g(n)

    1  g(n) < f(n)  < c2 g(n)

      n     < 2n+6 < 4n
      
 Use O notation  to express number of primitive operation executed as function of input size.
 
 #### Asymptotic Notation (terminology):
 
     * Logarithimic : O(logn)
     * Linear       : O(n)
     * Quadratic    : O(n^2)
     * Polynomial   : O(n^k) k-> constant and k>=1
     * Exponential  : O(a^n) a>1
     
     
## The "Big Omega" Ω-Notation:

Big - Omega notation is used to define the lower bound of an algorithm in terms of Time Complexity.

That means Big - Omega notation always indicates the minimum time required by an algorithm for all input values. That means Big - Omega notation describes the best case of an algorithm time complexity.

Big - Omega Notation can be defined as follows..

Consider function f(n) the time complexity of an algorithm and g(n) is the most significant term. If f(n) >= C x g(n) for all n >= n0, C > 0 and n0 >= 1. Then we can represent f(n) as Ω(g(n)).
     
<img src="Big Omega Graph.png",width=400,height=200>


## The "Big- Theta" Notation:

Big - Theta notation is used to define the average bound of an algorithm in terms of Time Complexity.

That means Big - Theta notation always indicates the average time required by an algorithm for all input values. That means Big - Theta notation describes the average case of an algorithm time complexity.

Big - Theta Notation can be defined as follows.

Consider function f(n) the time complexity of an algorithm and g(n) is the most significant term. If C1 g(n) <= f(n) >= C2 g(n) for all n >= n0, C1, C2 > 0 and n0 >= 1. Then we can represent f(n) as Θ(g(n)).

## Abstract Data Type

The idea of an ADT is to separate the notions of specification (what kind of thing we're working with and what operations can be performed on it) and implementation (how the thing and its operations are actually implemented).

**The benefits of using ADTs include:**

Code is easier to understand (e.g., it is easier to see "high-level" steps being performed, not obscured by low-level code).

** Type of Operation **:
    
    * Constructor:  You can create an instance of particular data type
    
    * Access Function: this let us access of particular data type.
    
    * Manipulation Procedure: this let us manipulation of particular data type.

** Why we need ADT in DS **:

    * Identify the requirement, what are the building block and what the requirement of that building block.
    * provide a language to talk on higher level of abstraction like in terms of stack , Queue
    * ADT encapsulate algorithm (how the data structure organize) and data structure that implement them
    * Seperate the issue of correctness and efficiency.


**Interested in Abstract:**

Specification like

    * What the instance would be like
    * what would be operation that permitted on these instances
    * what would be the axiom that governs those operation that tell us what operation we are doing.

### Stack:

Stack is collection of object that are inserted and removed according to last in first out(LIFO) manner.

Object is inserted at any time but only the last object can be removed.

**pushing** -> Inserting an item

**Popping** -> removing an item

#### stack is an ADT that support four main method***

  **New()**: Create a new stack
  
  **push(S:ADT and o, object)**: Insert object o on top of stack S
  
  **pop(S:ADT)**: Remove object from top of stack S,if stack is empty an error occured.
  
  **top(S:ADT)**: Return top object of stack, without removing it, If stack is empty an error occured.
  
 #### Other Method 
 
   **SIZE(S:ADT)**: return the number of object in stack S
   
   **IsEmpty(S:ADT)**: Indicate if Sstack is empty.
   
 #### Axioms:
 
   **pop(push(s,v))**=s

## STACK Implementation

````javascript
function stack(){
    this._size=0;
    this._storage=[];
}

stack.prototype.peek=function(){
    return this._storage[this._size];
}

stack.prototype.isEmpty=function(){
    //console.log('this._size',this._size);
    if(this._size>0) 
        return true; 
    else 
        return false;
}


stack.prototype.push=function(data){
    this._size=this._size+1;
    this._storage[this._size]=data;
}

stack.prototype.pop=function(){
    var size=this._size,
        deleteData;
    deleteData=this._storage[this._size];
    //console.log('this.isEmpty()',this.isEmpty());
    if(this.isEmpty()){
        delete this._storage[size];
        this._size=this._size-1;
    }else{
        console.log('empty stack')
    }
    return deleteData;
}

var S=new stack();
S.push(1);
S.push(2);
S.push(3);
console.log(S.peek());
console.log(S._storage);
S.pop();
console.log(S._storage);
S.pop();
console.log(S._storage);
````

## Queue:

A Queue is different form stack in that its removal and insertion follow FIFO (fist in fist out) principle.

Element are inserted in rear and removed from front.

**Three Fundamental methods are: ***

   **New()**: create empty queue.
   
   **enqueue(S:ADT,o:Object)**: insert object o at the rear of Queue S.
   
   ** dequeue(S:ADT)**: Remove object from the front of Queue ; an error occured if the queue is empty.
   
   **front(S:ADT)** : Return , but does not remove, front element; An error occured if queue is empty.
   
** Other Methods **

   **SIZE(S:ADT)**
   
   **ISEmpty(S:ADT)**
   
## Queue Implemenation

````javascript
var queue=function(){
    this._newestIndex=0;
    this._oldestIndex=0;
    this.storage=[];
}
queue.prototype.enqueue=function(item){
    if(item){
            this.storage[this._newestIndex]=item;
            this._newestIndex=this._newestIndex+1;
        }
    else{
        console.log('data is not correct');
    }
}
queue.prototype.dequeue=function(){
    if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        var deletedItem;
        deletedItem =this.storage[0];
        delete this.storage[0];
        this._oldestIndex=this._oldestIndex+1;
        console.log('Item is popup from Queue is '+ deletedItem);
    }
}
queue.prototype.front=function(){
     if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        console.log('Item is on front Queue is '+ this.storage[this._oldestIndex]);
    }
}

queue.prototype.size=function(){
    return (this._newestIndex - this._oldestIndex)
}

queue.prototype.printData=function(){
    var firstIndex=this._oldestIndex;
    if(this._oldestIndex === this._newestIndex){
        console.log('Queue is empty');
    }
    else{
        while(firstIndex !== this._newestIndex){
            console.log('Item:',this.storage[firstIndex])
            firstIndex=firstIndex+1;
        }
    }
}

var queue=new queue();
queue.front();
queue.dequeue();
queue.enqueue(1);
queue.front();
queue.enqueue(2);
queue.enqueue(3);
queue.dequeue();
queue.front();
console.log('size',queue.size());
queue.printData();
````

## Circular Singly linklist

````javascript
var Node=function(){
    this.item=null;
    this.next=null;
}
var CSLinklist=function(){
    this.head=null;
    this.size=0;
    this.tail=null;
}

//Insert in Cicular linklist

CSLinklist.prototype.add=function(item,position){
    if(item){
        var node=new Node();
        node.item=item;
        node.next=null;
        if(this.head===null && this.size < 1){
            this.head=node;
            this.tail=node;
            this.size=this.size+1;
        }
        else if(position===undefined || this.size < position){
            
            this.tail.next=node;
            node.next=this.head;
            this.tail=node;
            this.size=this.size+1;
        }else{
            if(position===1){
                node.next=this.head;
                this.head=node;
                this.tail=this.head;
                this.size=this.size+1;
            }
            else{
                var count=1;
                var current=this.head;
                var prev=current;
                while(count!==position){
                    count=count+1;
                    prev=current;
                    current=current.next;
                }
                prev.next=node;
                node.next=current;
            }
            this.size=this.size+1;
        }
    }
    else{
        console.log('Item is not defined')
    }
}


CSLinklist.prototype.SearchAt=function(position){
    if(this.head !==null){
        if(position===this.size){
            console.log('Item at position ',position ,'is ',this.tail.item);
        }else if(position > this.size){
            console.log('No Item of linklist at this position');
        }
        else{
            var count=1;
            var current=this.head;
            while(count !== position){
                count=count+1;
                current=current.next;
            }
            console.log('Item at position ',position, 'is ',current.item);            
        }
    }
    else{
        console.log('linklist is empty');
    }
}

CSLinklist.prototype.removeAt=function(position){
         if(position){
        if(position>this.size){
            position=this.size;
            console.log('position is greater than size so I am deleting last element')
        }
        if(position===1){
            this.head=this.head.next;
            this.tail.next=this.head;
            this.size=this.size-1;
        }else{
            var count=1;
            var current=this.head;
            while(count!==position-1){
                current=current.next;
                count=count+1;
            }
            if(position===this.size){
                current.next=this.head;
                this.tail=current;
            }else{
                current.next=(current.next).next;
            }
            this.size=this.size-1;
        }
        
        
    }  else{
        console.log('position is not defined');
    }
}

CSLinklist.prototype.print=function(){
        if(this.size>0){
        var current=this.head;
            while(current!==this.tail){
                console.log('item is: ',current.item);
                current=current.next;
            }
        console.log('item is: ',current.item);
        }else{
            console.log('Linklist is Empty');
        }
}

var Clink=new CSLinklist();
Clink.add('1');
Clink.add('2');
Clink.add('3');
Clink.add('4',2);
Clink.add('5');
Clink.add('6');
Clink.add('7');
Clink.add('8',8);

Clink.SearchAt(1);
Clink.SearchAt(8);
Clink.SearchAt(7);
Clink.removeAt(1);
Clink.print();
````
